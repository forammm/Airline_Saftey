# Part-2: Working with Models 

# 1)  Perform encoding in columns Gender, Customer_Type, Type_of_Travel, and Class.  

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab = LabelEncoder()
```

```python
df['Gender'] = lab.fit_transform(df['Gender'])
```

```python
df['Customer_Type'] = lab.fit_transform(df['Customer_Type'])
```

```python
df['Type_of_Travel'] = lab.fit_transform(df['Type_of_Travel'])
```

```python
df['Class'] = lab.fit_transform(df['Class'])
```

```python
df.shape
```

```python
df.info()
```

# 2)  Drop the column id

```python
df = df.drop(['id'],axis=1)
```

```python
df.shape
```

# 3)  Create the features and target Data

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df.drop('satisfaction',axis=1)
y = df['satisfaction']
```

```python
X.shape
```

# 4)  Perform scaling on features data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scal = scaler.fit_transform(X)
```

# 5)  Fit the decision tree model with various parameters  

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

# 6)  Split the data in training and testing sets  

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scal,y,test_size=0.4)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred  = model.predict(X_test)
y_pred
```

# 7)  Create a function to display precision score, recall score, accuracy, classification report, confusion matrix, F1 Score. 

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score
```

```python
matrix = confusion_matrix(y_test,y_pred, labels=[0,1])
print('Confusion matrix : \n',matrix)
```

```python
from sklearn import tree
text_representation = tree.export_text(model)
print(text_representation)
```

```python
print("Accuracy:" ,accuracy_score(y_test,y_pred))
```

```python
print("classification_report:\n",classification_report(y_test,y_pred))
```
